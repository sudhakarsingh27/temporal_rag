```python
import os
import requests
from bs4 import BeautifulSoup

# Base URL for the Beige Book reports
base_url = "https://www.federalreserve.gov/monetarypolicy/beigebook2000.htm"

# Function to download a PDF file
def download_pdf(url, folder, filename):
    response = requests.get(url)
    if response.status_code == 200:
        with open(os.path.join(folder, filename), 'wb') as f:
            f.write(response.content)
        print(f"Downloaded {filename}")
    else:
        print(f"Failed to download {filename}")

# Function to scrape and download PDFs
def scrape_and_download_pdfs(url, start_year, end_year):
    # Create a folder to store the PDFs
    if not os.path.exists("beige_books"):
        os.makedirs("beige_books")
    
    # Loop through the years
    for year in range(start_year, end_year + 1):
        # Update the URL for each year
        year_url = url.replace("2000", str(year))
        response = requests.get(year_url)
        if response.status_code != 200:
            print(f"Failed to retrieve data for {year}")
            continue
        
        # Parse the HTML content
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Find all the PDF links
        pdf_links = soup.find_all('a', href=True)
        for link in pdf_links:
            if link['href'].endswith('.pdf'):
                pdf_url = "https://www.federalreserve.gov" + link['href']
                filename = pdf_url.split('/')[-1]
                download_pdf(pdf_url, "beige_books", filename)

# Main function
if __name__ == "__main__":
    scrape_and_download_pdfs(base_url, 2000, 2025)
```

```python
import os
import requests
from bs4 import BeautifulSoup

def download_pdfs(start_year, end_year, output_dir):
    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)
    
    # Iterate over the years
    for year in range(start_year, end_year + 1):
        BASE_URL = f"https://www.federalreserve.gov/monetarypolicy/beigebook{year}.htm"
        # Fetch the webpage
        response = requests.get(BASE_URL)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')


        year_section = soup.find('th', text=str(year))
        if not year_section:
            print(f"No data found for the year {year}")
            continue

        # Find all PDF links for the year
        table = year_section.find_parent('table')
        links = table.find_all('a', href=True, text='PDF')

        for link in links:
            pdf_url = link['href']
            if not pdf_url.startswith('http'):
                pdf_url = 'https://www.federalreserve.gov' + pdf_url

            pdf_name = pdf_url.split('/')[-1]
            pdf_path = os.path.join(output_dir, pdf_name)

            # Download the PDF
            try:
                print(f"Downloading {pdf_name}...")
                pdf_response = requests.get(pdf_url)
                pdf_response.raise_for_status()

                with open(pdf_path, 'wb') as pdf_file:
                    pdf_file.write(pdf_response.content)
                print(f"Saved {pdf_name} to {output_dir}")
            except Exception as e:
                print(f"Failed to download {pdf_name}: {e}")

if __name__ == "__main__":
    START_YEAR = 2000
    END_YEAR = 2025
    OUTPUT_DIR = "beige_book_pdfs"

    download_pdfs(START_YEAR, END_YEAR, OUTPUT_DIR)
```
